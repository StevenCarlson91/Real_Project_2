DOE
I chose Latin Hypercube with 30 samples for my DOE. Initially I encountered errors while trying to run it, so I narrowed the possible range for the rotor thickness, outer diameter, and inner diameter. The rotor thickness ranged from 20-25 millimeters, the outer diameter was 130-150 millimeters, and the inner diameter was 60-80 millimeters. The minimum equivalent stress was calculated to be 1.08231E+07 Pa, and the maximum was 1.1782E+07 Pa. For the modal analysis, the minimum was 1123.5 Hz, and the max was 1495.8 Hz. The minimum temperature was 310.55 and the max was 332.02 degrees Celsius.





Goodness of Fit

I used the standard response surface  full second order polynomial fit and used 4 verification points. My model is simple, and the predicted vs observed data follows a similar trend, however, the maximum and average error arent that low. 

Local Sensitivity

The sensitivity analysis shows somewhat expected results. All 3 variables have a roughly equal impact on the maximum stress. The frequency is most highly impacted by the rotor outer diameter, and the temperature relies almost entirely on the rotor thickness.

Optimization
My goal here was to minimize the equivalent maximum stress and the temperature. I used the screening method with 100 samples, and it produced 3 candidate points.

Predictably, increasing the diameter of the rotor leads to a smaller stress and temperature.

Unfortunately, it appears that the outer and inner diameters didnt converge into a trend, so its hard to tell how much impact these had on the optimization process. Its possible that the range in inner and outer diameter was too limiting to have much of an impact.


The 3 candidate points are reasonable, but perhaps I played it too safe with the variable ranges to get better fitting values. Also, the screening optimization method should be used as an initial test, but my setup took several hours to complete, so I was hoping for better results.
